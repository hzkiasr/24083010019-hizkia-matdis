**3. Jumlah kata dalam kalimat**

```python
# Pisahkan teks menjadi kalimat
sentences = re.split(r'[.!?]+', clean_text)
sentences = [s.strip() for s in sentences if s.strip()]

# Hitung jumlah kata per kalimat
word_counts = [len(re.findall(r'\b[A-Za-zÀ-ÿ]+\b', s)) for s in sentences]

# menyimpan ke DataFrame
df = pd.DataFrame({
    'Kalimat': [f'Kalimat {i+1}' for i in range(len(sentences))],
    'Jumlah_Kata': word_counts
})

print("Jumlah Kata per Kalimat")
print(df)

#  visualisasi histogram
plt.figure(figsize=(7,4))
plt.hist(word_counts, bins=range(1, max(word_counts)+2), color='skyblue', edgecolor='pink', alpha=0.7)
plt.title("Distribusi Jumlah Kata per Kalimat")
plt.xlabel("Jumlah Kata")
plt.ylabel("Frekuensi Kalimat")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(f" Rata-rata panjang kalimat: {sum(word_counts)/len(word_counts):.2f} kata.")

```

**4. Jumlah kalimat dalam paragraf**

```python

#Pisahkan teks menjadi paragraf
paragraphs = [p.strip() for p in text.strip().split('\n\n') if p.strip()]

# Hitung jumlah kalimat dalam tiap paragraf
def count_sentences(paragraph):
    # Pisah berdasarkan tanda titik, tanda seru, atau tanda tanya
    sentences = re.split(r'[.!?]+', paragraph)
    sentences = [s.strip() for s in sentences if s.strip()]
    return len(sentences)

sentence_counts = [count_sentences(p) for p in paragraphs]

df = pd.DataFrame({
    'Paragraf': [f'Paragraf {i+1}' for i in range(len(paragraphs))],
    'Jumlah_Kalimat': sentence_counts
})

print(df)


# Buat histogram
plt.figure(figsize=(7,4))
plt.bar(df['Paragraf'], df['Jumlah_Kalimat'], color='pink', edgecolor='red')
plt.title("Jumlah Kalimat per Paragraf")
plt.xlabel("Paragraf")
plt.ylabel("Jumlah Kalimat")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
